社融总量的增减不能简单的看数据拐点来判断趋势，只有深入到数据拐点背后发生的变化才能进行推断。  
社融包括供给和需求两个维度，如果是政策收紧导致供给减少，那么社融大概会领先经济6~9个月下降  
如果是需求下降导致社融减低，那么社融和经济基本上是同时下降的  
如果仅仅是社融出现变化，而背后的经济和政策没有发生变化，那么这些变化可能仅仅是季节性的变化或者是数据上的结果

## 第二章、历史平均收益和未来收益 

用平均收益和波动性来衡量资产会产生系统性的偏差，因为波动率不能完全反映风险，比如流动性带来的风险。流动性差的资产会带来流动性溢价，但是实际上流动性差的资产由于交易量低，波动率很低，它的风险主要是来自于在危机发生时收到的冲击更大，所以流动性差的资产一般观察到的夏普比更高   
在实证的回报中，起点和终点的选择对于长期回报的计算结果影响非常大，但是如果把时间窗口拉的太长又会遇到早期数据质量过差和相关性低、出现结构性变化的问题

在用历史数据进行回测研究收益率时，一般需要把数据的起点和终点都放在估值水平相对中枢或者比较一般的位置上才有意义M

估值的提升是一个非常有意思的问题，收益一方面来自于资产带来的的利润回报，另一方面来自于资产的估值提升，理论上如果我们认为资产的估值有一个长期的中枢，那么预期收益应该不包含估值的变化预期，估值变化带来的收益是一个意外的收益来源。事实上，估值的提升会带来实现收益率的提升，但是其实是会降低预期收益率，这一点在债券市场里面特别明显。

如果预期收益是随时间变化的，那么简单的历史数据的复盘对于预期收益的判断就没有任何意义，但是书中认为一些前瞻性的指标比如价值指标和利差指标可以帮助我们判断预期收益,这种思路就是寻找一定的锚点

## 第三章、历史业绩：过去20年的长期视角

从更长期的视角来看（50年），国债的收益率随久期并不是单调增长的，应该是先上升后下降的  
对于信用债而言结果也是类似的，收益率随着评级的下降也是先上升后下降的  

## 第四章、术语介绍

预期收益率随时间可能是变化的，历史平均以实现收益反应的是 历史平均预期收益+历史平均非预期收益。根据估值水平的变化调整历史平均收益率的做法只能提供我们关于过去平均预期收益率的估计。  
预期收益和市场预期的收益率不一致。比如在股市连续上涨的时期中，人们预期的收益率会上升，但是其实由于估值的提升，预期收益应该下降。  
外国风险资产的套期保值收益率接近于投资者本国货币存款利率和外国资产相对外国存款利率的超额收益之和。

凯利公式: $ f=\frac{b\times p-q}{b} $  
式中f为你该用资产多少比例下注  
b为盈亏比  
p为胜率  
q为亏损概率，即q=1-p

1 期望值为正时，凯利公式是在赌徒免于破产的情况下，最快速增加资产的仓位控制。

2 期望值为零与负时，停止下注。

3 相同期望值时，提高系统的胜率可以提高最大仓位，提高资产增长率。

4 凯利公式应用于股票和期货市场时，由于市场状态的不同，而不能使用过于激进的凯利公式计算仓位。

5 通过改进或者降低凯利公式，将其应用于股票和期货市场。

## 第九章、债券风险溢价

债券的风险溢价指的是长期债券相对于短期无风险债券的收益率溢价，又称为期限溢价，投资期溢价，到期溢价或者久期溢价。   
债券的风险溢价（预期期限溢价）和实现的平均超额债券收益率和收益率曲线陡峭程度不同。     
债券风险溢价是对无风险收益率不确定性的补偿  
10年期利率 = 未来10年预期平均1年期利率+债券风险溢价  
$Y_{10} = E(平均Y_{1})+BRP_{Y}$  
$Y_{10} = E(平均通胀率)+E(平均R_{1})+BRP_{Y}$  
曲线陡峭化程度：  
$BRP_{Y}$ 是债券预期未来每年$BRP_{H}$ 的平均值  
$YC = Y_{10} - Y{1} $   
   $= [E(平均Y_{1})-Y{1}]+BRP_{Y}   $   
   $= E(\Delta 平均Y_{1}) +BRP_{Y}$
   $= 久期_{10}*E(\Delta Y_{10}) +BRP_{Y}$   
   
债券的市场收益率实际上就反映了当时的名义收益率预期，所以本书的核心在于如何从名义收益率预期中提取出风险溢价

关于YC，学界有两种观点，一个是认为完全是预期影响，另一个是认为是风险溢价，本书认为均值回归的利率预期对收益率曲线的影响更大

从长期的历史数据来看，收益率曲线是曲棍球形状，当期限相对短的时候债券的夏普比最高，短期限债券的平均夏普比特别高又叫做前端债券密度特别大，产生这一现象的原因是配置1年期左右债券的投资者很少，市场割裂

### 债券风险溢价的预期衡量指标

1. 收益率曲线陡峭度（YC)  
最常用，缺点是里面还包含了市场对于未来利率变化的预期  
实证研究发现YC和超额下个季度/下一年度的债券超额收益有一定关系，并且和未来短期限利率的变化有一定的正相关关系（这其实也好理解，当收益率曲线陡峭时，均值回归思想下一半会预期短期利率上行）
2. 根据远期利率曲线预测的实证债券风险溢价估计值（C-P BRP)  
试图剔除利率预期成分得到单纯的风险溢价。  
该指标和收益率曲线曲率有一定的关系。  
把收益率对1、2、3、4、5年期的远期利率回归，发现斜率系数具有简洁对称的帐篷形状，在2-3年的远期利率处达到最大值   
远期利率曲线越弯曲，长期债券的收益率越高  
3. 从期限模型中提取 K-W BRP
4. 基于调查数据的债券风险溢价衡量指标SBRP

从实证数据上看，YC的波动很大程度上是预期收益率波动引起的，通胀相关的风险溢价其实波动相对比较小

用YC来作为债券的预期风险溢价其实并不是一个非常好的指标，因为YC包含了利率变化的预期和风险溢价，利率变化的预期和风险溢价常反向变动从而使得YC只对短期限风险溢价有一定的预测效果，对长期的债券风险溢价没有好的预测。这是因为短期限利率有均值回归的特点，当短期限利率高的时候，市场有短期限利率下降的预期，这种情况下曲线一般比较平，不能反映风险溢价，更多反映短端利率的下降预期。利率曲线呈现出周期性的波动，很大程度上是因为人们均值回归的预期

### 债券风险溢价的来源

1. 通胀的不确定性是债券风险溢价的一个主要来源，而通胀的不确定性一般和通胀的水平有正相关性。   
在19世纪通胀十分低的时期，收益率曲线常常是负的，这是因为债券风险溢价非常低。进入21世纪，收益率曲线常常平坦化，这也是部分因为在明显的通胀锚下通胀预期不断下降，导致债券的风险溢价下降。

2.协方差风险和避险资产溢价。   
一般认为债券是相对安全的资产，可以用来做对冲，这样就会带来一个避险资产的溢价。 
从历史上看，股债之间的相关性并不稳定，股债相关系数的波动反映了债券作为避险资产的角色不断发生变化。
当需求冲击导致通胀和经济增长沿着菲利普斯曲线移动时，通胀和经济增长是正相关，此时股债负相关，债券起到一个对冲作用，即使给一个负的风险溢价也是合理的（因为和股票市场相关性是负的，所以CAPM模型给出债券的风险溢价是负的）。
当供给冲击导致菲利普斯曲线移动是，通胀和增长负相关，股债正相关，那么债券的风险溢价应该是正的。
这里需要注意沿着菲利普斯曲线移动是指预期通胀稳定，实际通胀变化。菲利普斯曲线的变动是由于预期通胀变动。
如果预期通胀稳定的情况下，股债就应该是负相关的。  
如果预期通胀变化，股债就是正相关的，这种情况下预期通胀的变化会对债券的预期收益产生多方面影响，一个是通胀不确定性溢价的变化，一个是避险资产性质减弱带来的的风险溢价变化。

3. 供求因素  
在2000年左右政府盈余时，国债供给短缺，收益率很高。书中认为养老金配置长期国债和亚洲央行对于长期国债的偏好导致风险溢价的下降   
研究发现当政府债务上升时，风险溢价也会上升

4.周期性因素

## 第十章、信用风险溢价

信用利差是指信用债超过同久期的国债的利率。但是信用利差不是信用溢价，信用利差包含了信用溢价、发行人的期权、预期违约补偿和流动性溢价。 
从历史数据的统计上看，信用债相对于国债已经实现的超额收益远小于信用利差，这主要是因为信用利差的变化，交易行为，降级偏差

影响信用利差的因素包括：  
1. 流动性溢价和国债的稀缺性
2. 周期性因素
3. 股票和利差的波动
4. 违约率

对于高评级债券，几乎没有违约风险，所以他的信用利差主要是反应了流动性因素和税收相关的因素。   
对于一般的债券（A级和BBB级），利差和周期性因素关系比较密切，当经济下滑时，利差往往增加。当股市波动率上升时，利差也上升。   
对于高收益债券，在经济危机发生时，违约率大增，利差急剧上升。   
违约损失 = 违约率*（1-回收率）

期限前段的信用债的收益风险比更高，这也与市场分割相关


如果把各种指标和下一时期信用债实现超额收益率进行相关性分析，可以发现公司债利差、国债/股市隐含波动率正相关，和经济活动呈现负相关
