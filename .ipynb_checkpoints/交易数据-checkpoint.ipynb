# 交易数据分析

分析：  
目标：找到数据中有什么规律，数据挖掘规律或者金融逻辑规律  
数据特征：不同时间、不同债券、不同投资主体三个维度下的净买入量
研究目标可以简化虽然有四个特征，但是我们关心的特征是净买入量
这里时间应该不能认为是一个特征，时间是一个物理上的维度，理论上和我们研究的问题没有理论上的关联，只能衡量先后顺序
有哪些可能规律？  
1.同一个record，有什么匹配模式？需要遍历四个特征的不同的组合，主要是看聚类问题  
2.不同的record之间有没有领先滞后关系  
3.不同的record之间有没有大小等可比关系  
守恒量和可变量  
同一天净买入量之和是0  
  
可能思考方向：  

如果不考虑当天利率变化：  
1. 机构间有没有一致性，比如某几个机构喜欢同时多/空  
2. 机构对于券有没有交易上的偏好？同一个券是否在不同机构有不同的交易比例，或者就喜欢多/空某个券  
3. 当出现多对一时，即多方空单一方多/多方多单一方空，谁在做独狼  
4. 某些券买/卖量就是大，某些机构买卖量就是大
5. 同一个券，不同机构连续买/卖？还是说买卖来回倒？
6. 有没有跟随效应，比如某个聪明机构先
7. 机构同一个券多翻空/空翻多时差
8. 不同机构出现大量买入/卖出时或者之后的特征

如果考虑当天利率变化：
1. 该券上涨/下跌时谁在多/空，机构-券-量 和上涨和下跌的关系
2. 不同机构交易比例大小和券上涨/下跌联系
3. 是否有聪明钱带领市场，说白了有没有先行指标
4. 放量交易后利率怎么走，无量交易后利率怎么走
5. 下跌时候谁在抄底，谁被鸽
6. 从净买入量看，什么是底部信号
7. 从净买入量看，什么是顶部信号
8. 那些券先涨或者先被买，那些后被买
9. 那些券先被卖，先咂那些券
10. 不同机构出现大量买入/卖出时或者之后的特征，异常增仓/减仓后的变化

如果多个券同时考虑，可以看同一机构当天总出货量和当天利率的关系，看看是不是有什么信号
